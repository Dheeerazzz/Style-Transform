```python
import tensorflow as tf
import cv2
import numpy as np 
import tensorflow_hub as hub
import matplotlib.pyplot as plt
```

```python
hub_model=hub.load('https://tfhub.dev/google/magenta/arbitrary-image-stylization-v1-256/2')
```

```python
def img_load(img_path):
    img=tf.io.read_file(img_path)
    img=tf.image.decode_image(img,channels=3)
    img=tf.image.convert_image_dtype(img,tf.float32)
    img=img[tf.newaxis,:]
    return img 
```

```python
content=img_load('content/content3.jpg')
style=img_load('arts/art5.jpg')
```

```python
plt.imshow(np.squeeze(content))
plt.show()
```

```python
plt.imshow(np.squeeze(style))
plt.show()
```

```python
styled_img=hub_model(tf.constant(content),tf.constant(style))[0]
```

```python
plt.imshow(np.squeeze(styled_img))
plt.show()
```
